```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe() 
```

```python
df = df[ df['Publishing Year'] > 1900]  #since the publishing year is in minus wee reframe the df for containing the dataset from 1900 onwards
```

```python
df.describe()
```

```python
df.isna().sum() #find the columns with total no. of null values
```

```python
df.dropna(subset = "Book Name" , inplace = True) #delete the rows with null values
```

```python
df.duplicated().sum() #to check any suplicated rows
```

```python
df.nunique() #to check the unique values 
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of publishing year")
plt.show()
```

```python
df['genre'].value_counts()
```

```python
#styling a graph using sns and matplotlib

sns.set(style="whitegrid")  
colors = sns.color_palette("Set2")  

# Plot
ax = df['genre'].value_counts().plot(
    kind='bar',
    color=colors,
    figsize=(8, 4),
    edgecolor='black'
)

# Set fonts and labels
plt.xlabel("Genre", fontsize=12, fontweight='bold', fontfamily='serif')
plt.ylabel("Number of Books", fontsize=12, fontweight='bold', fontfamily='serif')
plt.title("Number of Books in Each Genre", fontsize=14, fontweight='bold', fontfamily='serif')

# Rotate x-labels if needed
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Show grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")  
colors = sns.color_palette("Set2")  

sns.boxplot(x="genre" , y="Book_ratings_count" , data = df , palette=colors)

# Set labels and title
plt.xlabel("Genre", fontsize=12, fontweight="bold")
plt.ylabel("Book Ratings Count", fontsize=12, fontweight="bold")
plt.title("Distribution of Book Ratings per Genre", fontsize=14, fontweight="bold")
```

```python
plt.scatter(df["sale price"] , df["units sold"])
plt.xlabel("sale price")
plt.ylabel("units sold")
plt.title("scatterplot of sale price vs units sold")
plt.show()
```

```python
language_counts = df['language_code'].value_counts().sort_values(ascending=True)
```

```python
language_counts
```

```python
plt.pie(language_counts , labels = language_counts.index , startangle = 90 , autopct = "%1.1f%%")

'''' startangle=90	Rotates the pie chart to start from the top (90 degrees). Makes it more readable.
autopct="%1..1f%%"	Tells matplotlib to show percentages on the chart.
⚠️ However, this has a typo! It should be: "%1.1f%%" (with a single dot). '''

plt.title("language distribution of books")

plt.show()
```
